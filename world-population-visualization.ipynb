```python
import plotly.graph_objs as go
import plotly.express as px
from plotly.offline import init_notebook_mode,iplot,plot
init_notebook_mode(connected=True)
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/countries-of-the-world/countries of the world.csv")
df.head()
```

```python
df.info()
```

```python
data = dict(
        type = 'choropleth',
        colorscale = 'Blues',
        locations = df['Country'],
        locationmode = "country names",
        z = df['Population'],
        text = df['Country'],
        colorbar = {'title' : 'Population countries of the world'},
      )
```

```python
layout = dict(title = 'Population countries of the world',
              geo = dict(projection = {'type':'orthographic'})
             )
```

The outpput will show on Jupyter Notebook. 
Please run this code on Jupyter Notebook. 
**Thanks!**

```python
choromap = go.Figure(data = [data],layout = layout)
plot(choromap)
```
